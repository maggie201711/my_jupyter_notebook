# 从腾讯网址 https://news.qq.com/zt2020/page/feiyan.htm 上找到各个省份的数据，进行数据分析

# 数据提取

```python
import requests
import time
stamp = time.time()

exact_url = 'https://view.inews.qq.com/g2/getOnsInfo?name=disease_h5&callback=&_=%d' % int(stamp * 1000)
# URL 里的callback是回调函数，一般是说请求回来之后我要如何处理，这里我们只要数据，不需要处理所以这块等号后边可以为空。
# URL 里的最后是时间戳 * 1000取整数
response = requests.get(exact_url)

#response = requests.get('https://news.qq.com/zt2020/page/feiyan.htm') # get 不需要用户名密码就可以访问


if response.status_code == 200:
    print('get the page')
else:
    print('not get the page')
    
print(response.text)

data = response.json() # json格式（打过好后是键和值）的字符串转换成python格式的字典
```

# 数据提取

```python
result= data['data']  # result的结果是：字符串里是字典
result
```

```python
import json
data_dict = json.loads(result)  # 字符串转字典
data_dict
```

```python
data_dict['chinaTotal']
```

```python
data_dict.keys()
```

```python
china_dict= data_dict['areaTree']
china_dict = china_dict[0]
province_data_list = china_dict['children']
print(province_data_list)
```

```python
#print(len(province_data_list))
province_full_dict = {}
for idx in range(len(province_data_list)):  # need a range
    provice_data = province_data_list[idx]
    province_full_dict[provice_data['name']] = provice_data['total']
```

```python
# list change to dict
import pandas as pd
df_china = pd.DataFrame(province_full_dict)
```

```python
df_china
```

```python
cities = list(df_china.head(0).astype(str))
print(list(df_china.head(0).astype(str)))
```

```python
confirm = list(df_china.loc["confirm"])
print(confirm)

#如果类型不能转换成list，可以不用.values而是用.astype(int)来指定类型
```

```python
from pyecharts.charts import Map
from pyecharts import options as opts
```

```python
pieces=[
    {'min':1, 'max':100},
    {'min':101, 'max':500},
    {'min':501, 'max':1000},
    {'min':1001, 'max':1500},
    {'min':1501, 'max':2500},
    {'min':2501}
    
]

c = (
    Map()
    .add('区域分布', [list(z) for z in zip(cities, confirm)], 'china')
    .set_global_opts(title_opts=opts.TitleOpts(title='疫情地图'),
                    visualmap_opts=opts.VisualMapOpts(is_piecewise=True, pieces=pieces))
)
c.render_notebook()

```
