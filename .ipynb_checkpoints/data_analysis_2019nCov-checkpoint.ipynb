从腾讯网址 https://news.qq.com/zt2020/page/feiyan.htm 上找到各个省份的数据，进行数据分析

## 数据读取

```python
# 数据的读取
from selenium import webdriver
browser = webdriver.Chrome('C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe')
# url要写全
url = 'https://news.qq.com/zt2020/page/feiyan.htm' 
browser.get(url)
html = browser.page_source
browser.close()
```

```python
print(html)
```

## 数据解析

```python
from bs4 import BeautifulSoup

# 以lxml解析html
mysoup = BeautifulSoup(html, 'lxml')

china_list = mysoup.find('div', class_='chianList')  # find 输出第一个匹配结果，即find_all()[0]
```

```python
print(china_list)
```

```python
# use find_all() 方法，输出全部的匹配结果，搜索当前tag的所有tag子节点,并判断是否符合过滤器的条件
# Detail refer: https://www.cnblogs.com/nyist-xsk/p/7609968.html
province_list = china_list.find_all('div', class_='placeItemWrap') 
```

```python
print(province_list)
```

```python
data_list = []
for province in province_list:
    name = province.find('h2',class_='blue').get_text()
    added = province.find('div',class_=['add ac_add', 'add ac_add ']).get_text()
    confirm = province.find('div', class_='confirm').get_text()
    heal = province.find('div', class_='heal').get_text()
    dead = province.find('div', class_='dead').get_text()
    data_list.append({
        'name':name,
        'added':added,
        'cofirm':confirm,
        'heal':heal,
        'dead':dead
    })
print(data_list)
```
